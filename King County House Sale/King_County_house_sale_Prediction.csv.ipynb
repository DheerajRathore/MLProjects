```python
features = ['bedrooms', 'bathrooms','floors','waterfront', 'view', 'condition','sqft_above',
       'sqft_basement','yr_built','lat', 'zipcode','long','grade','sqft_living15', 'sqft_lot15']
target = df.price
X = df[features]
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test, Y_train,y_test = train_test_split(X,target,test_size=0.2,random_state=10) 
```

```python
#creating linera model.
from sklearn.linear_model import LinearRegression
linearModel = LinearRegression()
linearModel.fit(X_train, Y_train)
```

```python
#predicting score of linear model.
linearModel.score(X_test, y_test)
```

```python
# Measuring accuracy of model using cross validation
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score

cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)
cross_val_score(LinearRegression(), X, target, cv=cv)
```

```python
#Using decision  regression tree

from sklearn.tree import DecisionTreeRegressor

decisionRegressor = DecisionTreeRegressor()
decisionRegressor.fit(X_train, Y_train)
```

```python
decisionRegressor.score(X_test, y_test)
```

```python
decisionRegressor.predict(X_test)
```

```python
y_test
```

```python
# using Random forest model for predicting the house price in king county
from sklearn.ensemble import RandomForestRegressor
randomForestRegressor = RandomForestRegressor(max_depth = 20,n_estimators=100)
randomForestRegressor.fit(X_train, Y_train)
```

```python
randomForestRegressor.score(X_test, y_test)
```
